```python
#creating a function to scrape text from a list
def scrape_text(web_el):
    list1=[]
    for i in web_el:
        try:
            
            list1.append(i.text.replace("\n",","))
        except Exception as e:
            print("exception in scrape_text",e)
            list1.append("--")
    return list1
```

```python
#Function to extract website name
def get_website():
    website=driver.find_elements_by_xpath("//span[@class='name-only-text']")
    time.sleep(1)
    return website
```

```python
#Function to extract airline names
def get_airline():
    airline=driver.find_elements_by_xpath("//span[@class='codeshares-airline-names']")
    time.sleep(1)
    return airline
```

```python
#Fucntion to extract departure and arrival times
def get_times():
    times__=[]
    times=driver.find_elements_by_xpath("//span[@class='time-pair']")
    #time.sleep(1)
    #for i in times:
     #   if(i!=''):
      #      times__.append(i)
    return times
```

```python
#Function to get airport names
def get_airports():
    airport__=[]
    airport=driver.find_elements_by_xpath("//div[@class='bottom-airport']")or("//div[@class='bottom-airport flag']")
    #time.sleep(1)
    #for i in airport:
     #   if(i!=''):
      #      airport__.append(i)
    return airport
```

```python
#Function to extract Number and name of stops
def get_stops():
    stops__=[]
    stops=driver.find_elements_by_xpath("//div[@class='section stops']")
    #time.sleep(1)
    #for i in stops:
     #   if(i!=''):
      #      stops__.append(i)
    return stops
```

```python
#Function to extract duration of journey
def get_duration():
    duration__=[]
    duration=driver.find_elements_by_xpath("//div[@class='section duration allow-multi-modal-icons']")
    #time.sleep(1)
    #for i in duration:
      #if(i!=''):
       #     duration__.append(i)
    return duration
```

```python
#Function to extract price
def get_price():
    price=[]
    prices=driver.find_elements_by_xpath("//div[@class='above-button']")
    for i in range(0,len(prices),2):
        price.append(prices[i])
    return price
```

The above code is to determine the lengths of all the features scraped which are added to the lists below only if the lengths are identical i.e price_,duration_,stops_ etc. have equal lengths while time_ and airports_ have double the length of the price_.This is because the same web element for arrival and departure was used.

```python
def remove_null(list1):
    list2=[]
    for i in list1:
        if(i==''):
            continue
        else:
            list2.append(i)
    return list2
```
